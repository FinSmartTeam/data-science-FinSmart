Hal ini mengindikasikan adanya Multikolinearitas (multicollinearity) yaitu fenomena di mana dua atau lebih fitur independen (variabel penjelas) dalam dataset memiliki hubungan linear yang sangat kuat atau saling berkolerasi tinggi.

Selanjutnya akan dilakukan cek  multicollinearity


```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 1. Ambil hanya kolom numerik (fitur independen)
# PENTING: Jangan masukkan kolom target 'Tipe' atau kolom ID!
X = df.select_dtypes(include=[np.number])

# 2. Buat DataFrame untuk menampung hasil VIF
vif_df = pd.DataFrame()
vif_df["Fitur"] = X.columns

# 3. Hitung skor VIF untuk setiap kolom
vif_df["Skor VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]

# 4. Tampilkan hasil diurutkan dari VIF tertinggi
print("===== HASIL PEMERIKSAAN MULTIKOLINEARITAS (VIF) =====")
print(vif_df.sort_values(by="Skor VIF", ascending=False).to_string(index=False))
```

Hasil pengujian menunjukkan nilai Variance Inflation Factor (VIF) yang jauh di atas batas aman ($VIF \ge 10$).

Skor diatas menandakan adanya masalah multikolinearitas berat, di mana informasi antar-kolom mentah saling tumpang tindih secara masif. Untuk mengatasi redundansi data tersebut, kami melakukan Feature Engineering berupa agregasi variabel ke dalam 4 pilar fungsional, yaitu Income, Needs, Wants, dan Savings. Langkah reduksi dimensi ini wajib dilakukan agar data menjadi lebih ringkas dan bersih, sehingga model machine learning nantinya dapat mempelajari pola keuangan dengan stabil, akurat, serta efisien saat melakukan inferensi.

**Machine Learning**

```python
# =========================================================
# MACHINE LEARNING
# RANDOM FOREST CLASSIFICATION
# =========================================================

# =========================================================
# 28. IMPORT LIBRARY MACHINE LEARNING
# =========================================================

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (

    accuracy_score,
    f1_score,
    classification_report,
    confusion_matrix

)
```

```python
# =========================================================
# 29. MENENTUKAN FEATURE DAN TARGET
# =========================================================

# =========================================================
# FEATURE (X)
# =========================================================
# Menggunakan nominal/agregasi mentah
# untuk menghindari data leakage

X = df[[

    'Income',
    'Needs',
    'Wants',
    'Savings',


]]

# =========================================================
# TARGET (y)
# =========================================================

y = df['Tipe']
```

```python
# =========================================================
# 30. LABEL ENCODING TARGET
# =========================================================

label_encoder = LabelEncoder()

y_encoded = label_encoder.fit_transform(y)

print("\n===== LABEL ENCODING =====")

for i, label in enumerate(label_encoder.classes_):

    print(f"{label} --> {i}")
```

# **Pemodelan**

```python
# =========================================================
# 31. TRAIN TEST SPLIT
# =========================================================

X_train, X_test, y_train, y_test = train_test_split(

    X,
    y_encoded,

    test_size=0.2,

    random_state=42,

    stratify=y_encoded

)

print("\n===== SHAPE DATA =====")

print("X_train :", X_train.shape)
print("X_test  :", X_test.shape)

print("y_train :", y_train.shape)
print("y_test  :", y_test.shape)
```

```python
# =========================================================
# 32. MEMBUAT MODEL RANDOM FOREST
# =========================================================

rf_model = RandomForestClassifier(

    n_estimators=200,

    max_depth=10,

    min_samples_split=5,

    min_samples_leaf=2,

    random_state=42

)
```

```python
# =========================================================
# 33. TRAINING MODEL
# =========================================================

rf_model.fit(X_train, y_train)
```

```python
# =========================================================
# 34. PREDIKSI DATA TRAIN
# =========================================================

y_train_pred = rf_model.predict(X_train)
```

```python
# =========================================================
# 35. PREDIKSI DATA TEST
# =========================================================

y_test_pred = rf_model.predict(X_test)
```

```python
# =========================================================
# 36. EVALUASI DATA TRAIN
# =========================================================

train_accuracy = accuracy_score(

    y_train,
    y_train_pred

)

train_f1 = f1_score(

    y_train,
    y_train_pred,

    average='weighted'

)

train_error = 1 - train_accuracy

print("\n====================================")
print("EVALUASI DATA TRAIN")
print("====================================")

print(f"Accuracy Train : {train_accuracy:.4f}")

print(f"F1-Score Train : {train_f1:.4f}")

print(f"Error Train    : {train_error:.4f}")
```

```python
# =========================================================
# 37. EVALUASI DATA TEST
# =========================================================

test_accuracy = accuracy_score(

    y_test,
    y_test_pred

)

test_f1 = f1_score(

    y_test,
    y_test_pred,

    average='weighted'

)

test_error = 1 - test_accuracy

print("\n====================================")
print("EVALUASI DATA TEST")
print("====================================")

print(f"Accuracy Test : {test_accuracy:.4f}")

print(f"F1-Score Test : {test_f1:.4f}")

print(f"Error Test    : {test_error:.4f}")
```

```python
# =========================================================
# 38. CEK OVERFITTING
# =========================================================

print("\n====================================")
print("CEK OVERFITTING")
print("====================================")

selisih_accuracy = train_accuracy - test_accuracy

print(f"Selisih Accuracy : {selisih_accuracy:.4f}")

if selisih_accuracy > 0.10:

    print("Model cenderung OVERFITTING")

else:

    print("Model cukup stabil")
```
